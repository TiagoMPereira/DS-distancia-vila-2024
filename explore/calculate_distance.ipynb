```python
import pandas as pd
from geopy import distance
```

## Importando dados

```python
data = pd.read_csv("../data/team-stadium-full.csv", index_col=0)
data
```

## Calculando distância até a Vila Belmiro

```python
vila_coord = tuple(data.loc[data["team"] == "Santos FC", ["lat", "long"]].values[0])
nabi_coord = tuple(data.loc[data["team"] == "RB Bragantino", ["lat", "long"]].values[0])
print(vila_coord)
print(f"Distância Vila <-> Nabi: {distance.distance(vila_coord, nabi_coord).km} km")
```

```python
def calc_vila_distance(coordinates: tuple):
    vila_coord = (-23.9511157, -46.3389913)
    return distance.distance(vila_coord, coordinates).km
```

```python
data["coordinates"] = [(d['lat'], d['long']) for _, d in data.iterrows()]
data["vila_distance"] = [calc_vila_distance(coord) for coord in data['coordinates']]

data
```

```python
data.to_csv("../data/distances_vila.csv", index = False)
```
